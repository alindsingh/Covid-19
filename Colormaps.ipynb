# Rainbow and Grey Colormaps


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Step 1: Load the dataset
df = pd.read_csv("C:\Datasets\covid_19_india.csv")
states = df['State/UnionTerritory']
confirmed = df['Confirmed']
# Creating a simple scatter plot to apply colormaps
x = np.arange(len(states))
# Step 2: Visualize using a rainbow colormap
plt.figure(figsize=(10, 6))
sc = plt.scatter(x, confirmed, c=confirmed, cmap='rainbow', s=100)
plt.colorbar(sc, label='Confirmed Cases')
plt.xticks(ticks=x, labels=states, rotation=45)
plt.xlabel('State')
plt.ylabel('Confirmed Cases')
plt.title('COVID-19 Confirmed Cases by State (Rainbow Colormap)')
plt.tight_layout()
plt.show()
# Step 3: Visualize using a luminescent-like colormap (Greys)
plt.figure(figsize=(10, 6))
sc = plt.scatter(x, confirmed, c=confirmed, cmap='Greys', s=100)
plt.colorbar(sc, label='Confirmed Cases')
plt.xticks(ticks=x, labels=states, rotation=45)
plt.xlabel('State')
plt.ylabel('Confirmed Cases')
plt.title('COVID-19 Confirmed Cases by State (Greys Colormap)')
plt.tight_layout()
plt.show()
```
