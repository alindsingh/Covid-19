# Visualizing Covid - 19 Data




## Comparing cured cases vs deaths due to Covid - 19 from March 2020 to September 2021

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load the dataset
df = pd.read_csv("C:\Datasets\covid_19_india.csv")
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
# Aggregate the data by 'Date'
aggregated_data = df.groupby('Date')[['Cured', 'Deaths']].sum().reset_index()
# Line Chart
plt.figure(figsize=(6, 3))
plt.plot(aggregated_data['Date'], aggregated_data['Cured'], label='Cured',
color='green', marker='o')
plt.plot(aggregated_data['Date'], aggregated_data['Deaths'], label='Deaths',
color='red', marker='x')
plt.title('COVID-19 Cured vs. Deaths Over Time in India')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Bar Chart for Total Numbers
totals = aggregated_data[['Cured', 'Deaths']].sum()
plt.figure(figsize=(6, 3))
totals.plot(kind='bar', color=['green', 'red'])
plt.title('Total COVID-19 Cured vs. Deaths in India')
plt.xlabel('Category')
plt.ylabel('Total Number of Cases')
plt.xticks(ticks=np.arange(2), labels=['Cured', 'Deaths'], rotation=0)
plt.tight_layout()
plt.show()
```
