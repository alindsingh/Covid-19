# Visualizing COVID-19 Data in India Using Pandas and Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
# Step 1: Load the dataset (Replace this with the actual file)
df = pd.read_csv("C:\Datasets\covid_19_india.csv")
# Step 2 & 3: Consider quantitative attributes & Remove missing values
df.dropna(subset=['Confirmed', 'Deaths'], inplace=True)
# Step 4: Apply horizontal and vertical alignment - Using barh for horizontal
df_sorted_confirmed = df.sort_values('Confirmed', ascending=False)
plt.figure(figsize=(10, 8))
plt.barh(df_sorted_confirmed['State/UnionTerritory'],
df_sorted_confirmed['Confirmed'], color='blue')
plt.xlabel('Confirmed Cases')
plt.ylabel('State')
plt.title('COVID-19 Confirmed Cases by State')
plt.tight_layout()
plt.show()
# Step 5: Apply ascending and descending ordering - Using regular bar for deaths
df_sorted_deaths = df.sort_values('Deaths', ascending=True)
plt.figure(figsize=(10, 8))
plt.bar(df_sorted_deaths['State/UnionTerritory'], df_sorted_deaths['Deaths'],
color='red')
plt.ylabel('Deaths')
plt.xlabel('State')
plt.title('COVID-19 Deaths by State')
plt.tight_layout()
plt.show()
```
